# **Tokenization of paragraphs/sentences**

---



```
import nltk


paragraph = """ A.P.J. Abdul Kalam (Fullname: Avul Pakir Jainulabdeen Abdul Kalam) served as the 11th president of the Republic of India between 2002 and 2007. He was an inspiring teacher, and a narrative author, and a great scientist.. """
               
# Tokenizing sentences
sentences = nltk.sent_tokenize(paragraph)

# Tokenizing words
words = nltk.word_tokenize(paragraph)
print(sentences)
```

# **Stop Words**

---





```
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```
print(stopwords.words('english'))

text = " A.P.J. Abdul Kalam (Fullname: Avul Pakir Jainulabdeen Abdul Kalam) served as the 11th president of the Republic of India between 2002 and 2007. He was an inspiring teacher, and a narrative author, and a great scientist.. "
text_tokens = word_tokenize(text)

tokens_without_sw = [word for word in text_tokens if not word in stopwords.words('english')]

print(text_tokens)
print(tokens_without_sw)
```

# **Stemming**

---



```
from nltk.stem.porter import *

porterStemmer = PorterStemmer()

sentence="NLP is the new buzz word of the present era. It has added colours to lives of many new age entrepreneurs as a tool for personal mastery."
wordList = nltk.word_tokenize(sentence)

stemWords = [porterStemmer.stem(word) for word in wordList]

print(' '.join(stemWords))
```

# **Word Lemma**

---




```
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "NLP is the new buzz word of the present era. It has added colours to lives of many new age entrepreneurs as a tool for personal mastery."
punctuations="?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word)))
```

```
from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
txt = """NLP is the practice of understanding how people organise their thinking, feeling, language and behaviour to produce the results they do."""
[wnl.lemmatize(i,j[0].lower()) if j[0].lower() in ['a','n','v'] else wnl.lemmatize(i) for i,j in pos_tag(word_tokenize(txt))]
```
